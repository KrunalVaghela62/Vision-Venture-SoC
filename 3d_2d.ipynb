```python
def world_to_pixel(x,y,z, K):

    # Convert pixel to normalized camera coordinates
    world_coords = np.array([x,y,z,1])  # Homogeneous image coordinates
    normalized_coords = K @ world_coords  # Apply inverse intrinsic matrix

    # Scale normalized coordinates by depth to get camera coordinates
    camera_coords = normalized_coords/z  # [X_c, Y_c, Z_c]




    # Extract world coordinates
    u,v,w = camera_coords[:3]
    return u,v
```

```python
import numpy as np
K=np.array([[646.35693359375,0,645.4840698242188,0],[0,645.5247192382812,367.06915283203125,0],[0,0,1,0]])
points=np.array([[-0.062,-0.079,0.397],[0.218,-0.074,0.388],[-0.066,0.133,0.384],[0.211,0.133,0.374]])
```

```python
result=[]
for x,y,z in points:
  u,v=world_to_pixel(x,y,z,K)
  result.append([u,v])
result=np.array(result)
print(result)
for x,y in result:
  print(x,y)
```
